```python
plt.figure(figsize = (24, 240))
plt.barh(range(len(col_names)), missing_rate, align = 'edge', color = '#969696')
plt.grid(alpha = 0.4)
plt.yticks(range(len(col_names)), col_names, fontsize = 4)
plt.xlabel('Missing Ratio',fontsize = 16)
plt.ylabel('Feature Name',fontsize = 16)
plt.title('Missing Rate Summary',fontsize = 32)
plt.savefig('missing_rate.pdf', dpi = 300)
```
